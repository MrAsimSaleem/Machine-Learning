```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("homeprices_onehotencoding.csv")
df
```

```python
# Getting dummies of a colum
dummies =pd.get_dummies(df.town)
dummies
```

```python
# Concatenating two df's column wise
df_merge = pd.concat([ df, dummies], axis = 'columns')
df_merge
    
```

```python
# Droping some columns town columns and one of these dummy colums
```

```python
final_data = df_merge.drop(["town", "robinsville"], axis = 'columns' )

# If we don't drop it sklearn model still will work perfectly
# but it is good to drop these items
```

```python
final_data
```

```python
# Lets do some thing like to train this model by linear model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# value of x calcualtion is based on the value of independent variables
# here  prices is dependent on other variables 
# so x
x = final_data.drop('price', axis = 'columns')
x
```

```python
# NOw its time for Y
y = final_data.price
y
```

```python
model.fit(x, y)
```

```python
# Now  play with prediction thing
# prices in west windsor town
model.predict([[2300, 0, 1]])
```

```python
# prices in 	monroe township
model.predict([[2400, 1, 0]])
```

```python
# Prices in robinsville town
model.predict([[2400, 0, 0]])
```

```python
# Lets check our model accuracy
model.score(x, y)
# 95% accuracy is in this model
```
