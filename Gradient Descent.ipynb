## Excersise

```python
def predict_using_sklearn():
    df = pd.read_csv("grad_exercise.csv")
    r = LinearRegression()
    r.fit(df[['math']],df.cs)
    return r.coef_, r.intercept_


def gradient_desc(x_math , y_cs):
    m = c =0
    n = len(x_math)
    iterations = 10000
    lear_rate = 0.003       # by shifting learning rate we can come close to optimal value
    cost_previous = 0
    for i in range(iterations):
        y_pred = m * x_math - c
#         Lets add cost function so that we could guess less cost range
        cost = 1 / n * sum([val **2 for val in (y_cs - y_pred)])
        der_wrt_m = -2/n * sum(x * (y_cs - y_pred))
        der_wrt_c = -2/n * sum(y_cs - y_pred)
        c = c- lear_rate * der_wrt_c
        m = m- lear_rate * der_wrt_m
        
        if math.isclose(cost, cost_previous, rel_tol=1e-20):
            break
        cost_previous = cost
        print("m = {}, c = {},cost = {}, iterations {}".format(m, c, cost, i))
    return m, c
```

```python
if __name__ == "__main__":
    df = pd.read_csv("grad_exercise.csv")
    x_math = np.array(df.math)
    y_cs = np.array(df.cs)


    m, c = gradient_desc(x_math, y_cs)
    print("Using gradient descent function: Coef {} Intercept {}".format(m, c))

    m_sklearn, c_sklearn = predict_using_sklearn()
    print("Using sklearn: Coef {} Intercept {}".format(m_sklearn, c_sklearn))
```

## Saving models


```python
df = pd.read_csv("homePrices.csv")
df
# df.reshape(1, -1)
```

```python
# Now its ready to be predicted as linear data
model= linear_model.LinearRegression()
model.fit(df[['area']], df.price)
print("coeff=" , model.coef_)
print("Intercept =",model.intercept_)
```

```python
model.predict([[2300]])
```
